```python
textFile = open("tweetsFile.txt")
textFile = textFile.readlines()
df = pd.DataFrame(columns=['ID','Text']).set_index('ID')
```

```python
allWords = []
i =0
for t in textFile:
    
    #remove header
    if t == "id,date,text\n":
        continue
    
    
    ID = int(t[0:19])
    text = t[40:(t.rfind(' '))]
    
    lastWord = t[(t.rfind(' '))+1:]
    
    if len(lastWord.strip()) != 0 and d.check(lastWord.strip()) == True:
        text += " " + lastWord
        
    if text[0:2] == 'RT':
        text = (text[2:]).strip()
        
    text = re.sub('[@#].*?\s', '',text)
    
    allWords.append([ID,text])
    i+=1
    if i%1000 == 0:
        print(i)
```

```python
df = pd.DataFrame(allWords,columns=['ID','tweets'])
```
